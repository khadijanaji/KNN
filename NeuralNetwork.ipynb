```python
def generateData():
    # generate random numbers between 0-1
    seed(1) # get the same data from random
    min=0
    max=5
    dataset=[]
    for _ in range(100):
        valueX1 = random() # real value between 0 and 1
        scaledvalueX1 = min + (valueX1 * (max - min))
        valueX2 = random() # real value between 0 and 1
        scaledvalueX2 = min + (valueX2 * (max - min))
        x1=round(scaledvalueX1,1) # one value after comma
        x2=round(scaledvalueX2,1) # one value after comma
        y0=f(x1)
        cl=0
        if(x2>y0):
            cl=1
        dataset.append([x1,x2,cl])
    return dataset
```

### 2- Comment peut on choisir une cost function ?
L'utilisation de l'agorithme Gradient Descent dans un réseau de neurone sera en premier lieu par choisir quel type de condition fonctionne, et utiliser la dérivée de cette fonction pour calculer les prochaines valeurs dans chaque iteration jusqu'au trouver la petite valeur de la fonction cost.

```python
# 3- Soit le programme #program2 donné en annexe

lr=0.001
n_iters=100000
def fit(X,y):
    bias=0
    W=np.array([0.,0.])
    n_samples,n_features=X.shape
    for i in range(n_iters):
        linear_model=np.dot(X,W)+bias
        y_predicted=sigmoid(linear_model) 
        dW=(1/n_samples)*np.dot(X.T,(y_predicted-y))
        dbias=(1/n_samples)*np.sum(y_predicted-y) 
        W-=lr*dW #ligne12 
        bias-=lr*dbias #ligne13 
    return bias,W
```

### 3-2- Quelles sont les modifications à apporter au code source pour minimiser Mean Absolute Error


```python
lr=0.001
n_iters=1000
def fitMAE(X,y):
    bias=0
    W=np.array([0.,0.])
    n_samples,n_features=X.shape
    for i in range(n_iters):
        linear_model=np.dot(X,W)+bias
        y_predicted=sigmoid(linear_model)
        # mean absolute error
        mae=(1/n_samples)*np.sum(np.abs(y-y_predicted))
        # dw
        dW=(-1/n_samples)*np.dot(X.T,y_predicted*(1-y_predicted))
        # dbias
        dbias=(-1/n_samples)*np.sum(y_predicted*(1-y_predicted))
        W-=lr*dW #ligne12 
        bias-=lr*dbias #ligne13 
    return bias,W
fitMAE(X,y)
```

### 3-4- Quelles sont les modifications à apporter au code source pour minimiser MSE !!

```python
lr=0.001
n_iters=1000
def fitMSE(X,y):
    bias=0
    W=np.array([0.,0.])
    n_samples,n_features=X.shape
    for i in range(n_iters):
        linear_model=np.dot(X,W)+bias
        y_predicted=sigmoid(linear_model)
        # Mean Squared Error
        mae=(1/n_samples)*np.sum((y-y_predicted)**2)
        # dw
        dW=(-2/n_samples)*np.dot(X.T,y_predicted*(1-y_predicted)*(y-y_predicted))
        # dbias
        dbias=(-2/n_samples)*np.sum(y_predicted*(1-y_predicted)*(y-y_predicted))
        W-=lr*dW #ligne12 
        bias-=lr*dbias #ligne13 
    return bias,W

fitMSE(X,y)
```

```python
def calcul_accuracy(y,y_pred):
    same=0
    all=len(y)
    for i,j in zip(y,y_pred):
        if i==j:
            same=same+1
    return same/all
```

* b. Loss value (binary_crossentropy)
* c. W, bias

```python
lr=0.001
n_iters=1

def fit(X,y):
    bias=0
    W=np.array([0.,0.])
    accuracy=[]
    n_samples,n_features=X.shape
    for i in range(n_iters):
        linear_model=np.dot(X,W)+bias
        y_predicted=sigmoid(linear_model)
        accuracy=calcul_accuracy(y,predict(bias,W,X))
        log_loss=(-1/n_samples)*np.sum(y*np.log(y_predicted)+(1-y)*np.log(1-y_predicted))
        dW=(1/n_samples)*np.dot(X.T,(y_predicted-y)) 
        dbias=(1/n_samples)*np.sum(y_predicted-y) 
        W-=lr*dW #ligne12 
        bias-=lr*dbias #ligne13 
    return bias,W,log_loss,accuracy
```

```python
bias,W,log_loss,accuracy=fit(X,y)
print("Iteration :{}".format(n_iters),bias,W)
```
