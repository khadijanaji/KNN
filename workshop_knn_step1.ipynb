```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sbrn
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.info()
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
df['Species'].value_counts()
```

```python
sbrn.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sbrn.stripplot(x='Species',y='SepalLengthCm',data=df,size=8,orient='v')
```

```python
sbrn.pairplot(df, hue="Species", palette="husl", markers=["o", "s", "D"])
```
