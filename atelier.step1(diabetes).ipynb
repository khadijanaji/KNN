```python
def matrice_confusion(y_true, y_pred):
    matrice=[[0,0],[0,0]]
    for i,j in zip(y_true,y_pred):
        if i==0 and j==0:
            matrice[0][0]+=1
        elif i==0 and j==1:
            matrice[0][1]+=1
        elif i==1 and j==0:
            matrice[1][0]+=1
        else:
            matrice[1][1]+=1
    
    return np.array(matrice)

matrice_confusion(y_real,y_pred)
```

```python
def accuracy(matrix):
    return (matrix[0][0]+matrix[1][1])/(np.sum(matrix))
accuracy(matrix)
```
