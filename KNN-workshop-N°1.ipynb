### M√©thode 1 (Normalisation)

```python
def normalisation(X):
    return X.apply(lambda x: (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0)))
```

```python
X = normalisation(X)
X
```

### M√©thode 2 (Normalisation sans boucle)

```python
def normalize(X):
    x_min = min(X.iloc[:,0])
    x_max = max(X.iloc[:,0])
    f = (lambda x: (x - x_min) / (x_max - x_min))
    X.iloc[:,0] = f(X.iloc[:,0])

    x_min = min(X.iloc[:,1])
    x_max = max(X.iloc[:,1])
    f = (lambda x: (x - x_min) / (x_max - x_min))
    X.iloc[:,1] = f(X.iloc[:,1])

    x_min = min(X.iloc[:,2])
    x_max = max(X.iloc[:,2])
    f = (lambda x: (x - x_min) / (x_max - x_min))
    X.iloc[:,2] = f(X.iloc[:,2])

    x_min = min(X.iloc[:,3])
    x_max = max(X.iloc[:,3])
    f = (lambda x: (x - x_min) / (x_max - x_min))
    X.iloc[:,3] = f(X.iloc[:,3])

    return X
```

```python
X = normalize(X)
X
```

```python
#M√©thode 2 : Avec Boucle
def normalizeLoop(X):
    for i in range(len(X.iloc[0])):
        x_min = min(X.iloc[:,i])
        x_max = max(X.iloc[:,i])
        f = (lambda x: (x - x_min) / (x_max - x_min))
        X.iloc[:,i] = f(X.iloc[:,i])
    return X
    
```

```python
normalizeLoop(X)
```

### Standardisation en Python

```python
def standardisation(X):
    for i in range(len(X.iloc[0])):
        f = (lambda x: (x - np.mean(x)) / (np.std(x)))
        X.iloc[:,i] = f(X.iloc[:,i])
    return X
```

```python
standardisation(X)
```
